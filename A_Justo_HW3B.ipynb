$\textbf{BASIC  BIOLOGICAL BACKGROUND}$ 


The most common modeling of cell dynamics involves the changes in concentration of proteins, DNA and RNA. In a lot of cases, a given protein can be related to a given behavior. For example, an increase in lactase is a decision taken by cell in order to process and decide the optimal source of energy. In other cases, the level of a given protein is not directly related with a given process, but can be a signal for the activation of a secondary process in the cell. In any case, making a decision in a cell often implies changing the concentration of a given protein from a low value to a high value or viceversa. Typically, if a gen is involved (or the interaction among them and proteins) one talks about genetic networks which are able to implement a decision. If the main driver of the decision is the interaction between proteins and external signals, one talks about a metabolic network. However, a lot of the times both refer to the same idea, interaction of genes, proteins and external signaling (or flow) to implement a process. And the first step to implement a process is very often to take a decision.

Since concentrations are the key, one should remember that any chemical reaction  will go to equilibrium unless some sort of input is introduced. Therefore, there is no decision-making or life in thermodynamic equilibrium. A flow of mass or energy is needed. In a given set of interactions between DNA, RNA and different proteins there is a background of structures which provides the energy and machinery to produce something new (a protein) out of equilibrium. The most classic example is the reading and transcription from DNA to mRNA with RNA polymerase and from mRNA to protein in the ribosome. The machinery provided by DNA polymerase and ribosome is therefore necessary. However, DNA is not the only source of information, the influx of signal from outside the cell is also relevant and, in some cases, the flow of a given molecule from the exterior is directly a source of crucial mass, as you may already know from the biophysics II analysis of glycolysis.

Besides the energy or mass that, most of the time, is in the background (when for example, the concentration of ATP or the number of ribosomes is given as a constant or hidden in some parameters), most of the decision -making models involve non-linear processes. We have seen how this nonlinear dynamics is the outcome of MM dynamics plus the key phenomena of cooperativity that you studied also in detailed in biophysics II. Cooperativity to produce a given element however is not the only feature of most metabolic or genetic network. Most proteins degrade and this is normally included in the analysis. More importantly, the inhibition or activation of a particular DNA transcription is key and will be addressed in more detail here. 

Finally, in some cases a decision is taken upon the reception of a signal which is transmitted from the membrane. The signal comes normally in the form of a ligand $L$ (epinephrine anyone?) which binds to a receptor $R$ (protein in the membrane) forming a complex $RL$ that can activated a signaling pathway. One of the most common is the G-signaling pathway, where a G-protein $Gn$ splits in its components $\alpha$ and $\beta$ with $\alpha$ ($Ga$) being involved in other cellular responses by activating the production of different proteins and enzymatic processes. The basic structure of the pathway is  $R+L<->RL$, $Gn+RL->Ga+Gb+RL$, $Ga->Gd$, $Gd+Gb->Gn$ where normally the last part is very fast and can be slaved. The key feature is that an external signal $L$ can be modulated by controlling the concentration of $Ga$ at the spot. Once the process cascades (with the actions that an increase in $Ga$ induces), further control is possible.

$\textbf{BASIC MATHEMATICAL BACKGROUND} $

The key feature of dynamical system that one should remember is that fixed points are quite often attractors of the dynamics (not always). This means that concentrations go to a given set of values if you wait long enough and the system is unperturbed. Depending on the initial condition you end up in one fixed point or the other. Fixed points in the dynamics are points where there is no flux in the dynamical system. If you are there you stay there. In 1d, all dynamical system end up in fixed points. They attract all initial conditions. In 2d, attractors can be fixed points or oscillations (also called limit cycles). In 3D, you have seen that besides fixed points and limit cycles you can find chaotic behavior. Actually, in discrete systems where time is not continuous, chaos appears even in 1d, as Romualdo has shown you. The key mathematical element is to understand the phenomena of bifurcation where a given fixed point splits in two (or the opposite, one disappears by merger with another) or exchange stability. 

The $ \textbf{bifurcations of co-dimension one}$ are changes in the dynamics in phase space which are discontinuous when one parameter is changed. The three bifurcations of co-dimension 1 are the transcritical, the saddle-node and the pitch-fork bifurcation (pitchfork bifurcation can be either subcritical or supercritical). These bifurcation are effective changes in the behavior of the fixed point along a given line of phase space, therefore, they can be studied with a one dimensional system. Close to a bifurcation point you can always write the dynamics close to the fixed point in the 1d dynamical system in a characteristic (canonical) way. One for each of the three types of bifurcations.

For the case of transcritical bifurcation, the dynamics close to the bifurcation is completely equivalent to the following dynamics:

\begin{equation}
 \frac{dx}{dt}= rx -x^2
\end{equation}

For the case of saddle-node is equivalent to:
\begin{equation}
 \frac{dx}{dt}= r + x^2
\end{equation}

For the case of pitchfork (supercritical) is equivalent to:
\begin{equation}
 \frac{dx}{dt}= rx - x^3
\end{equation}

The key parameter in all cases is $r$. The reduction of the dynamics is performed so that $r$ is the distance to the bifurcation point, so that, by construction, the bifurcation happens at  $r=0$. The dynamics is completely different (and the fixed points too) as parameter $r$ goes from negative to positive. For example, for the transcritical bifurcation, $x=0$ and $x=r$ are clearly fixed points, but for $r<0$ it is stable and for $r>0$ is unstable. The dynamics will tend to go to $x=0$ for $r<0$ and suddenly change and go to $x=r$ for $r>0$. This is an exchange of stability. Similar analysis can be made for the saddle node and/or the pitchfork. In the saddle node for $r<0$ there are two fixed points $x=\sqrt{-r}$ (unstable) and $x=-\sqrt{-r}$ (stable) and for $r>0$ suddenly two fixed points disappear.  For the pitchfork $x=0$ is always a fixed point, but for $r<0$ there are no more fixed points and for $r>0$ there are $x=\sqrt r$ and $x=-\sqrt r$. In this case for $r<0$, the unique fixed point $x=0$ is stable, and for $r>0$ it becomes unstable while the other two become stable.


$\textbf{A basic genetic network. Decision via transcritical bifurcation} $

Any of the bifurcations can be used to take a decision. The idea behind this is the following, by changing a signal which acts as a parameter one changes the stability of the fixed point and changes the attractor of the dynamics. This is what we basically call making a decision. We change from one concentration (close to zero for example) to another one finite. The fact that in the saddle-node bifurcation, fixed points can annihilate implies that you can actually take a decision where a concentration which is fixed (at the fixed point) must suddenly change when a signal arrives if this signal destroys the fixed point. The idea of the examples and code is to visualize two examples among the different possibilities that use bifurcations.

We start with a basic genetic network where the decision is taken via transcritical bifurcation. The basic biological structure you know from biophysics II to produce a protein is a one-directional flow: from DNA, you get mRNA, and from mRNA you get the protein. To get the first basic network is actually rather easy. In DNA, close to the sites where RNA polymerase (RNAP) starts reading, there is normally another site called operator site where proteins can attach. The attachment of these proteins affect the rate at what RNAP binds to DNA. These proteins can regulate the transcription by either promoting or inhibiting its attachment. In other words, the operator site is an activator or a inhibitor of the production of mRNA which depends on the presence of another protein. This other protein is called a transcription factor. You have seen all this on Biophysics II.

We focus now on a transcription factor which activates transcription. Consider the production of mRNA as having a basic rate from the RNAP when the operator is not affected. mRNA ($m=[mRNA]$) degrades but produces a protein ($p=[P]$) that also degrades. Such a linear flow system would read:

\begin{equation}
 \frac{dm}{dt}= \alpha-\mu_m m \;\;\;\;\;\; \frac{dp}{dt}= \beta m-\mu_p p       
\end{equation}


On the other hand, if $K_O$ is the equilibrium constant of the binding of the transcription factor to the operator site, the fraction of operator sites with the protein bounded is $F=\frac{p/K_O}{1+p/K_O}$, where $p$ is the concentration of the protein that acts as a transcription factor. It is clear the transcription rate will be higher when the empty operators are more common. The transcription rate should go from $\alpha$ to $\alpha F$. With this ingredient, the simplest network that you can construct with an interaction is that a protein is its own activation transcription factor. This is, the presence of a protein produces it own transcription. The set of equations for this dynamical system is:

\begin{equation}
 \frac{dm}{dt}= \frac{\alpha p}{K_O+p} -\mu_m m \;\;\;\;\;\; \frac{dp}{dt}= \beta m-\mu_p p   \;\;\;\;\;(1)
\end{equation}

One can reduce this system if considers the mRNA dynamics very fast and slaves it. We get:

\begin{equation}
 \frac{dp}{dt}= \frac{\alpha \beta}{\mu_m}\frac{p}{p+K_O} -\mu_p p=A \frac{p}{p+K_O}-\mu_p p
\end{equation}

If you perform an analysis of the fixed points you can see that you have a transcritical bifurcation at the key parameter:

\begin{equation}
 S=\frac{A}{\mu_p K_O}=\frac{\alpha \beta}{\mu_m \mu_p K_O}=1.
\end{equation}

To explain the results obtained, one can rely on the analysis of the equilibria of the system.

\begin{equation}
 0= \frac{\alpha p}{K_O+p} -\mu_m m \;\;\;\;\;\; 0= \beta m-\mu_p p   \;\;\;\;\;
\end{equation}

\begin{equation}
 m=\frac{\mu_p}{\beta} p   \;\;\;\;\;
\end{equation}

\begin{equation}
 \frac{\alpha p}{K_O+p}=\mu_m \frac{\mu_p}{\beta} p \;\;\;\;\;\;
\end{equation}

So, one possibility is $p=0$, and there is also another possibility: $p=\frac{\alpha\beta}{\mu_m\mu_p}-K_0$. Therefore:

\begin{equation}
 p_{eq1}=0 \;\;\;\;\;\; m_{eq1}=0   \;\;\;\;\;
\end{equation}

\begin{equation}
 p_{eq2}=\frac{\alpha\beta}{\mu_m\mu_p}-K_0 \;\;\;\;\;\; m_{eq2}=\frac{\mu_p}{\beta}\left(\frac{\alpha\beta}{\mu_m\mu_p}-K_0\right)   \;\;\;\;\;
\end{equation}

So, knowing that there is a transcritical bifurcation at the parameter

\begin{equation}
 S=\frac{A}{\mu_p K_O}=\frac{\alpha \beta}{\mu_m \mu_p K_O}=1.
\end{equation}

one would expect the stable point of the system to be $p_{eq1}=0$, $m_{eq1}=0$ for $S<1$, and $p_{eq2}$, $m_{eq2}$ for $S>1$, which is the situation that is actually observed.

How the cell implements the change in $S$? The control of the degradation ($\mu_p$ or $\mu_m$) is very complex and out of the scope of this course, but it can certainly be done in the cell. Equally, the base rate  $\alpha$ can be changed too. This is the one you coded. How does the cell control $\alpha$? The most common way to do it is by making another protein or signal interact with the same operator that activates it. For example, a very large $\alpha_2$ can be partially inhibited by another operator to a base value $\alpha$ but when this inhibition disappears the value of the transcription increases to $\alpha_2$ crossing the bifurcation threshold. From class you should know how two proteins would interact with the same operator site. # ON/OFF decision making. A general model approach

To understand how a more drastic ON/OFF decision can be made in the cell, we will use the Collins-Toggle switch (see details if you wish in Nature 403, 339-342). The model is very easy to describe and write. A gene produces a protein $p_1$ which acts as inhibitor of the production of another protein $p_2$, and this second protein acts as inhibitor of the first protein. Both inhibitions act cooperatively with exponent $n=2$ and are regulated by two other external signals/proteins which can inhibit the inhibition. This is: $I_1$ inhibits the inhibition that $p_1$ produces on $p_2$, and $I_2$ does the same for the inhibition of $p_2$ on $p_1$. Slaving the mRNA of each process and assuming that both proteins degrade and are produced at the same base rate, we can write the system of equations as:

<br>

\begin{eqnarray}
 \frac{d p_1}{dt} &=& \frac{\alpha}{1+p_2^2/I_2^2}- \mu p_1   \\
 \frac{d p_2}{dt} &=& \frac{\alpha}{1+p_1^2/I_1^2}- \mu p_2    
\end{eqnarray}

<br>

More complex now is to understand the dynamics of this model. First notice that, in principle, $I_1$ and $I_2$ may have its own dynamics: This dynamics is what affects the decision making process. The way to approach this problem is to study how the fixed points change with the parameters $I_1$, $I_2$ and then to observe what kind of bifurcations are present upon changing these parameter

$\textbf{Visualize nullclines}$

Any analysis must start looking with the fixed points $ \frac{d p_1}{dt}=\frac{d p_2}{dt}=0$. It is easy to find that there is always a solution where $p_1=p_2=p_f$ if $I_1=I_2=I$ with $p_f$ being a cubic equation $p_f(I^2+p_f^2)-K_I=0$ with $K_I=\alpha I^2 / \mu$. It is more cumbersome to investigate the instability of these solutions without the help of symbolic analysis of the computer. Computing the jacobian matrix in the fixed point is complex, and studying the diagonalization is more complex. However, finding fixed points where $p_1 \neq p_2$ is even more difficult. The best way to proceed in this case is to use a numerical approach analyzing the nullclines.


When setting any of the derivatives of a dynamic variable to zero, one finds a constrain which allows you to establish an algebraic equation between all the variables. In phase space of dimension $d$ this leads to a subspace with dimension $d-1$. In the case of a 2d dynamical system, setting one derivative to zero gives a line in 2d space. For example in our previous model, setting to zero each of the derivatives one can write

<br>

\begin{eqnarray}
 f(p_1,p_2)=\frac{\alpha}{1+p_2^2/I_2^2}- \mu p_1=0   \\
 g(p_1,p_2)=\frac{\alpha}{1+p_1^2/I_1^2}- \mu p_2=0     \\
\end{eqnarray}

<br>

One can see that these are basically two lines by rearranging the previous equations:

\begin{eqnarray}
 p_2=I_2 \sqrt{\frac{\alpha}{\mu p_1}-1}  \\
 p_2 = \frac{\alpha}{\mu} \frac{1}{1+p_1^2/I_1^2}    \\
\end{eqnarray}

# Introducing dynamics in controlling parameters. No parameter anymore


The final ingredient is introducing a dynamics to $I_1$ and $I_2$ so as to observe this decision as $I_1$ and $I_2$ produce the corresponding bifurcations in the dynamical system. We implement this  first with a very simple dynamics and later using a G-signaling process which modifies the concentration of $I_2$. We will do the first in class, and you will have to do the second at home. We will not change $I_1$ in any case for simplicity.

Simulate the basic model with $I_1$ and $I_2$ frozen for a set of parameters where the final state is symmetric and another set of parameters where the protein presence is higher in one case than the other. In this case there are two fixed points and the final state depends on the initial condition. Proceed to set a simulation where  $I_2$ is unfrozen and its dynamics has a characteristic time scale larger than the one in the 2d switch. $I_2$ must evolve from an initial condition where the large $p_2$ fixed point exist to a concentration where it does not. You must clearly see at which point the decision is taken. It is clear then that an external signal which controls $I_2$ will immediately lead to a decision. 

In other words, if you want $I_2$ to go from $IA$ to $IB$. Pick the initial condition of $I_2(t=0)=IA$ and then add the dynamic equation:

\begin{equation}
 \frac{d I_2}{dt}=-K_I (I_2-IB).
\end{equation}

where $K_I$ sets a time scale which is larger than the characteristic time of $p_1$ and $p_2$ but short enough so that you can see it evolve.


The set of equations to be solved reads:

\begin{eqnarray}
 \frac{d p_1}{dt} &=& \frac{\alpha}{1+p_2^2/I_2^2}- \mu p_1   \\
 \frac{d p_2}{dt} &=& \frac{\alpha}{1+p_1^2/I_1^2}- \mu p_2   \\
 \frac{d I_1}{dt} &=& 0                                       \\
 \frac{d I_2}{dt} &=& -K_I (I_2-IB)
\end{eqnarray}

```python
%matplotlib inline
import matplotlib.pylab as plt
import math
import numpy as np

Tmax = 100
dt = 0.1
n = int(Tmax/dt)

# *********************** Both I1 and I2 frozen ***********************

p1o = 0.1; p2o = 0.1; I1o = 0.1; I2o = 0.1
alpha = 0.1; mu = 0.2;


jp1=np.zeros(n); jp2=np.zeros(n)
p1vec=np.zeros(n); p2vec=np.zeros(n); I1vec=np.zeros(n); I2vec=np.zeros(n); tvec=np.zeros(n)

for i in range(n):
    
    jp1[i] = alpha/(1+p2o**2/I2o**2)-mu*p1o
    jp2[i] = alpha/(1+p1o**2/I1o**2)-mu*p2o
    
    p1o=p1o+dt*jp1[i]
    p2o=p2o+dt*jp2[i]
    
    p1vec[i] = p1o
    p2vec[i] = p2o
    I1vec[i] = I1o
    I2vec[i] = I2o
    tvec[i]  = i*dt
    
plt.figure(figsize=(20, 9));
plt.subplot(1,2,1)
plt.plot(tvec,p1vec,label='$p_1$')
plt.plot(tvec,p2vec,label='$p_2$')
plt.plot(tvec,I1vec,label='$I_1$')
plt.plot(tvec,I2vec,label='$I_2$')
plt.xlabel('t (ms)', fontsize=18)
plt.ylabel('Concentrations (mM)', fontsize=18)
plt.legend( loc=4,prop={'size':18})
plt.title("Both $I_1$ and $I_2$ frozen", fontsize=18)
plt.xlim(0, Tmax); plt.ylim(0, 0.5)
plt.tick_params(labelsize=18)

# *********************** I1 frozen and I2 unfrozen ***********************

IA = 0.1; IB = 0.2; p1o = 0.1; p2o = 0.1; I1o = 0.1; I2o = IA
alpha = 0.1; mu = 0.2; KI = 0.001

jp1=np.zeros(n); jp2=np.zeros(n); jI2=np.zeros(n)
p1vec=np.zeros(n); p2vec=np.zeros(n); I1vec=np.zeros(n); I2vec=np.zeros(n); tvec=np.zeros(n)

for i in range(n):
    
    p1vec[i] = p1o
    p2vec[i] = p2o
    I1vec[i] = I1o
    I2vec[i] = I2o
    tvec[i]  = i*dt
    
    jp1[i] = alpha/(1+p2o**2/I2o**2)-mu*p1o
    jp2[i] = alpha/(1+p1o**2/I1o**2)-mu*p2o
    jI2[i] = -KI*(I2o-IB)
    
    p1o=p1o+dt*jp1[i]
    p2o=p2o+dt*jp2[i]
    I2o=I2o+dt*jI2[i]
    
plt.subplot(1,2,2)
plt.plot(tvec,p1vec,label='$p_1$')
plt.plot(tvec,p2vec,label='$p_2$')
plt.plot(tvec,I1vec,label='$I_1$')
plt.plot(tvec,I2vec,label='$I_2$')
plt.xlabel('t (ms)', fontsize=18)
plt.ylabel('Concentrations (mM)', fontsize=18)
plt.legend( loc=4,prop={'size':18})
plt.title("$I_1$ frozen and $I_2$ unfrozen", fontsize=18)
plt.xlim(0, Tmax); plt.ylim(0, 0.5)
plt.tick_params(labelsize=18)
```

## Coupling a decision process with to G-signaling



$\textbf{Compartment near the membrane}$

We are going to think as all the volume close to the membrane as a single compartment, representing the membrane where you want to model the levels of protein $G_a$. Its evolution must be taken from a two-variable model obtained from slaving the following system:

$$R + L  \underset{k_{a}}{\overset{k_{b}}{\longleftrightarrow}} RL$$

$$G + RL \xrightarrow[]{k_{Ga}} G_a + G_{bg} + RL$$

$$G_a \xrightarrow[]{k_{Gd}} G_d$$

$$G_d + G_{bg} \xrightarrow[]{k_{G1}} G$$


Notice that, regarding $G$-related proteins, there are three reactions with four different concentrations, namely $G$, $G_a$, $G_d$, $G_{bg}$. However, only two are independent since the combination $G+G_{bg}$ is always constant and $G+G_a+G_d$ also remains constant. In other words, $G+G_{bg}=GB_{tot}$ and $G+G_a+G_d=GA_{tot}$ where $GB_{tot}$ and $GA_{tot}$ are constants that are set when you provide initial condition for the system.  To simplify this, set the inital conditions such that $G_{bg}(t=0)=G_a(t=0)=G_d(t=0)=0$ so that $G(t=0)=GA_{tot}=GB_{tot}=G_{tot}$. The dynamical system for $RL$, $G$ and $G_a$ becomes:

\begin{equation}
 \frac{d [RL]}{dt} = -k_{b} [RL] +k_a(R_{tot}-[RL]) [L]
\end{equation}

\begin{equation}
 \frac{d G}{dt} = -k_{Ga} G [RL] +k_{G1}(G_{tot}-G) (G_{tot}-G-G_a)
\end{equation}

\begin{equation}
 \frac{d G_a}{dt} = k_{Ga} G [RL] -k_{Gd}G_a
\end{equation}


This dynamical system of three variables has some parameters that you do not know, check the notes (pdf) for some of them or look at the literature. You do not need exact values, just the order of magnitude. You can check in the pdf in atenea that $k_{G1}$ is orders of magnitude larger than any other rate . It should be very easy for you to see that you must slave $G$. This slaving is extremely easy to do if you consider that $k_{G1}$ is so large that $k_{Ga} [RL]/(k_{G1} G_{tot})$ is going to be always very close to zero. You do not need to test your slaving here..just implement it. In other words, slave $G$, and find the relation between $G$ and $G_a$ with the condition $k_{Ga} [RL]/(k_{G1}G_{tot}) \approx 0$. $\textbf{Coupling a decision process with $I_2$ to G-signaling}$

You must address now a more realistic process where a cell receives a signal, you process it in the membrane via the G-cascade, you generate protein $I_2$ close to the membrane, and $I_2$ then diffuses to the center where you have the Collins-Switch where a change in $I_2$ levels leads to an increase in the level of a given protein. In other words, you must simulate a more realistic decision-making process. Regarding this simulation, you must build a dynamical system of two compartments which are connected by diffusion of the protein $I_2$. The first compartment represents the membrane, the second one the cytosol.

The key idea is that $G_a$ helps to produce $I_2$, this is, $G_a$ produces the protein which inhibits the inhibition of the switch. The dynamical equation for the concentration of $I_2$ near the membrane should read:

\begin{equation}
 \frac{d I_{2m}}{dt} = P(G_a) -D_I (I_{2m} -I_{2c}) -\mu_I  I_{2m}
\end{equation}

where $P(G_a)$ is a production term which will depend on the level of the $G$ protein in active configuration $G_a$. Set yourself the term $P(G_a)$ as it would appear in a Michaelis-Menten reaction with cooperativity $n=2.5$. Observe that the equation for $I_{2m}$ also includes a degradation term plus a diffusion term which links the concentration of $I_2$ near the membrane ($I_{2m}$) with the concentration of $I_2$ in the cytosol $I_{2c}$. For the diffusion constant, look into bionumbers, and for the degradation rate take that its characteristic time scale is around a couple of minutes.


$\textbf{Second compartment: In the cytosol}$

$I_2$ in the cytosol ($I_{2c}$) is not constant. $I_2$ must be generated in the cytosol both from the production from the ribosome and from diffusion coming from the membrane. It must also degrade.

\begin{equation}
 \frac{d I_{2c}}{dt} = H(I_{2c}) + f_v D_I (I_{2m} -I_{2c}) -\mu_I  I_{2c}
\end{equation}

You must set the production of  $I_2$ from mRNA-Ribosome $H(I_{2c})=\frac{\alpha_I}{1+(I_{2c}/K_I)}$ so that it is self-inhibitory. This is, $I_2$ self-inhibits its own production. It is clear that you will have to pick $\alpha_I$ and $K_I$ to define $H$. You will have to play with these parameters until you get a proper response.

Take $f_v=0.01$, which represents a volume factor, as a given. This factor compensates the fact that the volume of the cytosol is much larger than the one near the membrane, therefore, a drop in the concentration of proteins in the membrane does not lead to the same increase in the concentration of protein in the cytosol.

Finally, you must use the Collins-Switch modified at will (proteins $p_1$ and $p_2$) in the cytosol, using $I_{2c}$ as the key inhibitory protein and observe that when $L$ increases in the membrane, at some point, the system goes from low to high level of one of the proteins. Regarding the other protein $I_1$, set it constant concentration everywhere, near the membrane and in the cytosol


All in all, the set of equations to be solved reads:

\begin{eqnarray}
 \frac{d [p_1]}{dt} &=& \frac{\alpha}{1+[p_2]^2/[I_{2c}]^2}- \mu [p_1]         \\
 \frac{d [p_2]}{dt} &=& \frac{\alpha}{1+[p_1]^2/[I_1]^2}- \mu [p_2]         \\
 \frac{d [I_1]}{dt} &=& 0                                             \\
 \frac{d [I_{2m}]}{dt} &=& \frac{\left(\frac{[G_a]}{K_d}\right)^{2.5}}{1+\left(\frac{[G_a]}{K_d}\right)^{2.5}} -D_I ([I_{2m}] -[I_{2c}]) -\mu_I  [I_{2m}]   \\
 \frac{d [I_{2c}]}{dt} &=& \frac{\alpha_I}{1+([I_{2c}]/K_I)} + f_v D_I \left([I_{2m}] -[I_{2c}]\right)-\mu_I  [I_{2c}] \\
 \frac{d [RL]}{dt} &=& -k_{b} [RL] +k_a(R_{tot}-[RL]) [L]             \\
 \frac{d [G]}{dt} &=& -k_{Ga} [G] [RL] +k_{G1}(G_{tot}-[G]) (G_{tot}-[G]-[G_a]) =0 \\
 \frac{d [G_a]}{dt} &=& k_{Ga} [G] [RL] -k_{Gd}[G_a]
\end{eqnarray}

Where $[G]$ is slaved. The outcome of that is, after straightforward calculation:

\begin{equation}
 [G]=G_{tot}-[G_a]
\end{equation}

Where $[G_a]$ is still given by the dynamics.

Qualitatively:
- Without any signal, both $p_1$ and $p_2$ tend to remain in a certain equilibrium.
- When signal $L_1$ is activated, then $p_2$ tends to be high and $p_1$ tends to be low.
- When signal $L_2$ is activated, then $p_1$ tends to be high and $p_2$ tends to be low.
- Those cases when one of the signals is activated and the other signal is activated with a few seconds delay, first is like the two previous cases, and then the high protein tends to decrease, whereas the low protein tends to increase until they reach a certain equilibrium (different equilibrium than the one regarding the case when no signals are created).

Note that, once any $[L]$ is activated, the corresponding $[RL]$ reaches its equilibrium, so even interrupting the creation of $[L]$, the concentration $[RL]$ remains high, thus not allowing the protein concentrations to return fast to their original state. This is because the degradation constant of the $[RL]$ is so low that degradation occurs at very long time scales. At the time scale of this experiment, practically no degradation of $[RL]$ is observed. If it occurred, that would imply that the proteins return to the first state where no signals were created.
